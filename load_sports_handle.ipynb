```python
import csv
import re
import pandas as pd
handleList = []
characters_to_remove = "()\n"
pattern = "[" + characters_to_remove + "]"
with open('sports.txt') as input_data:
    for line in input_data:
        if line.find("@") != -1:
            newLine = re.sub(pattern, "", line)
            row = newLine.split("@")
            handle = "@" + row[1]
            handleList.append(handle)
        else:
            continue
sports_df = pd.DataFrame({'handles':handleList})
sports_df.to_csv('sports_df.csv')
sports_df.head()
#sports_df.shape
```
