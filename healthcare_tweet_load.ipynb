```python
import csv
import re
import pandas as pd
handleList = []
characters_to_remove = "()\n"
pattern = "[" + characters_to_remove + "]"
with open('community_web_content.txt') as input_data:
    for line in input_data:
        if line.find("@") != -1:
            newLine = re.sub(pattern, "", line)
            row = newLine.split("@")
            handle = "@" + row[1]
            handleList.append(handle)
        else:
            continue
healthcare_df = pd.DataFrame({'handles':handleList})
healthcare_df.to_csv('healthcare_df.csv')
healthcare_df.head()
healthcare_df.shape
```

```python
import tweepy
import csv
import os
import pandas as pd

# since these are secrets, we have not kept it in the git hub repo, 
# so, please update the secret values for the following 
consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)

def get_tweets(screen_name):

    alltweets = []  
    
    new_tweets = tweepy.Cursor(api.user_timeline, screen_name = screen_name).items(500)
    alltweets.extend(new_tweets)

    outtweets = [[tweet.id_str, tweet.created_at, tweet.text, tweet.favorite_count, tweet.retweet_count, tweet.entities, tweet.lang] for tweet in alltweets]
    #write the csv  
    with open(f'{screen_name}_tweets.csv', 'w') as f:
        writer = csv.writer(f)
        writer.writerow(["id","created_at","text", "favorites", "retweets", "entities", "language"])
        writer.writerows(outtweets)
     
    pass
```

```python
handles = pd.read_csv('healthcare_df.csv')
names = handles['handles'].to_list()
```

```python
#1
if __name__ == '__main__':
    #loop through the handles in the list
    for i,name in enumerate(names):
        try:
            get_tweets(name)
        except Exception as e: print(e)
```
